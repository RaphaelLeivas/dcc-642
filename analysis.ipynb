```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 18}) # Set global font size to 18

minimax = pd.read_csv(f'minimax-alfabeta.csv')
minimax
```

```python
depth2_time = minimax[minimax['depth'] == 2]['time'] * 1000
depth3_time = minimax[minimax['depth'] == 3]['time'] * 1000
depth4_time = minimax[minimax['depth'] == 4]['time'] * 1000
depth5_time = minimax[minimax['depth'] == 5]['time'] * 1000

print(depth2_time.mean(), depth2_time.std())
print(depth3_time.mean(), depth3_time.std())
print(depth4_time.mean(), depth4_time.std())
print(depth5_time.mean(), depth5_time.std())

plt.figure(figsize=(10,6))

plt.hist(depth2_time, label='Depth = 2', edgecolor="black")
plt.hist(depth3_time, label='Depth = 3', edgecolor="black")
plt.hist(depth4_time, label='Depth = 4', edgecolor="black")
plt.hist(depth5_time, label='Depth = 5', edgecolor="black")

plt.xlabel('Tempo (ms)')
plt.ylabel('Frequência')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
depth2_nodes = minimax[minimax['depth'] == 2]['nodes']
depth3_nodes = minimax[minimax['depth'] == 3]['nodes']
depth4_nodes = minimax[minimax['depth'] == 4]['nodes']
depth5_nodes = minimax[minimax['depth'] == 5]['nodes']

print(depth2_nodes.mean(), depth2_nodes.std())
print(depth3_nodes.mean(), depth3_nodes.std())
print(depth4_nodes.mean(), depth4_nodes.std())
print(depth5_nodes.mean(), depth5_nodes.std())


plt.figure(figsize=(10,6))

plt.hist(depth2_nodes, label='Depth = 2', edgecolor="black", bins=5)
plt.hist(depth3_nodes, label='Depth = 3', edgecolor="black", bins=5)
plt.hist(depth4_nodes, label='Depth = 4', edgecolor="black", bins=5)
plt.hist(depth5_nodes, label='Depth = 5', edgecolor="black", bins=5)

# plt.xscale('log')
plt.xlabel('Nós expandidos')
plt.ylabel('Frequência')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
data_alfabeta = {'depth': [2,3,4,5],
        'time': [1.7,6.4,10.1,83.2],
        'nodes': [31,125.6,265,2366.4]}
df_alfabeta = pd.DataFrame(data_alfabeta)

data_minimax = {'depth': [2,3,4,5],
        'time': [3.6,14.1,92.5,618],
        'nodes': [48.2,384,2357,16769]}
df_minimax = pd.DataFrame(data_minimax)

# Create the plot
plt.figure(figsize=(10,6))

plt.plot(df_alfabeta['depth'], df_alfabeta['time'],'o-',  label='Alfa-Beta', color="red") 
plt.plot(df_minimax['depth'], df_minimax['time'], 'o-', label='Minimax', color="blue") 

# Add labels and a title
plt.xlabel('Depth')
plt.ylabel('Tempo médio (ms)')

# Add a legend
plt.yscale('log')
plt.legend()
plt.grid()

# Display the plot
plt.show()
```

```python
# Create the plot
plt.figure(figsize=(10,6))

plt.plot(df_alfabeta['depth'], df_alfabeta['nodes'],'o-',  label='Alfa-Beta', color="red") 
plt.plot(df_minimax['depth'], df_minimax['nodes'], 'o-', label='Minimax', color="blue") 

plt.xlabel('Depth')
plt.ylabel('Número de Nós Expandidos')

# Add a legend
plt.yscale('log')
plt.legend()
plt.grid()

# Display the plot
plt.show()
```

```python
ids = pd.read_csv(f'ids.csv')
ids
```

```python
limit1000_time = ids[ids['limit'] == 1000]['time'] * 1000
limit2000_time = ids[ids['limit'] == 2000]['time'] * 1000

print(limit1000_time.mean(), limit1000_time.std())
print(limit2000_time.mean(), limit2000_time.std())

plt.figure(figsize=(10,6))

plt.hist(limit1000_time, label='Limite = 1 s')
plt.hist(limit2000_time, label='Limite = 2 s')

# plt.xscale('log')
plt.xlabel('Tempo gasto (ms)')
plt.ylabel('Frequência')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
limit1000_nodes = ids[ids['limit'] == 1000]['nodes']
limit2000_nodes = ids[ids['limit'] == 2000]['nodes']

print(limit1000_nodes.mean(), limit1000_nodes.std())
print(limit2000_nodes.mean(), limit2000_nodes.std())

plt.figure(figsize=(10,6))

plt.hist(limit1000_nodes, label='Limite = 1 s')
plt.hist(limit2000_nodes, label='Limite = 2 s')

# plt.xscale('log')
plt.xlabel('Nós vistados')
plt.ylabel('Frequência')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
limit1000_depth = ids[ids['limit'] == 1000]['depth']
limit2000_depth = ids[ids['limit'] == 2000]['depth']

print(limit1000_depth.median(), limit1000_depth.std())
print(limit2000_depth.median(), limit2000_depth.std())

plt.figure(figsize=(10,6))

plt.hist(limit1000_depth, label='Limite = 1 s')
plt.hist(limit2000_depth, label='Limite = 2 s')

# plt.xscale('log')
plt.xlabel('Profundidade')
plt.ylabel('Frequência')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
