```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 18}) # Set global font size to 18

minimax = pd.read_csv(f'minimax.csv')
minimax
```

```python
depth2_time = minimax[minimax['depth'] == 2]['time'] * 1000
depth3_time = minimax[minimax['depth'] == 3]['time'] * 1000
depth4_time = minimax[minimax['depth'] == 4]['time'] * 1000
depth5_time = minimax[minimax['depth'] == 5]['time'] * 1000

print(depth2_time.mean(), depth2_time.std())
print(depth3_time.mean(), depth3_time.std())
print(depth4_time.mean(), depth4_time.std())
print(depth5_time.mean(), depth5_time.std())

plt.figure(figsize=(10,6))

plt.hist(depth2_time, label='Depth = 2', edgecolor="black", bins=5)
plt.hist(depth3_time, label='Depth = 3', edgecolor="black", bins=5)
plt.hist(depth4_time, label='Depth = 4', edgecolor="black", bins=5)
plt.hist(depth5_time, label='Depth = 5', edgecolor="black", bins=5)

plt.xlabel('Tempo (ms)')
plt.ylabel('Frequência')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
depth2_nodes = minimax[minimax['depth'] == 2]['nodes']
depth3_nodes = minimax[minimax['depth'] == 3]['nodes']
depth4_nodes = minimax[minimax['depth'] == 4]['nodes']
depth5_nodes = minimax[minimax['depth'] == 5]['nodes']

print(depth2_nodes.mean(), depth2_nodes.std())
print(depth3_nodes.mean(), depth3_nodes.std())
print(depth4_nodes.mean(), depth4_nodes.std())
print(depth5_nodes.mean(), depth5_nodes.std())


plt.figure(figsize=(10,6))

plt.hist(depth2_nodes, label='Depth = 2', edgecolor="black", bins=5)
plt.hist(depth3_nodes, label='Depth = 3', edgecolor="black", bins=5)
plt.hist(depth4_nodes, label='Depth = 4', edgecolor="black", bins=5)
plt.hist(depth5_nodes, label='Depth = 5', edgecolor="black", bins=5)

plt.xscale('log')
plt.xlabel('Nós expandidos')
plt.ylabel('Frequência')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
